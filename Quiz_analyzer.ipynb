```python
import plotly.graph_objects as go

# Sample data for visualizations
topics = ['Math', 'Science']
topic_correct = [8, 5]
topic_total = [10, 10]

difficulties = ['Easy', 'Medium']
difficulty_correct = [6, 4]
difficulty_total = [10, 10]

# Create an interactive bar chart for Performance by Topic
fig_topic = go.Figure()

fig_topic.add_trace(go.Bar(
    x=topics,
    y=topic_correct,
    name='Correct Answers',
    marker_color='blue'
))

fig_topic.add_trace(go.Bar(
    x=topics,
    y=topic_total,
    name='Total Questions',
    marker_color='lightgray'
))

fig_topic.update_layout(
    title='Performance by Topic',
    xaxis_title='Topics',
    yaxis_title='Number of Answers',
    barmode='group',
    hovermode='x unified'
)

# Show the figure
fig_topic.show()

# Create an interactive bar chart for Performance by Difficulty Level
fig_difficulty = go.Figure()

fig_difficulty.add_trace(go.Bar(
    x=difficulties,
    y=difficulty_correct,
    name='Correct Answers',
    marker_color='green'
))

fig_difficulty.add_trace(go.Bar(
    x=difficulties,
    y=difficulty_total,
    name='Total Questions',
    marker_color='lightgray'
))

fig_difficulty.update_layout(
    title='Performance by Difficulty Level',
    xaxis_title='Difficulty Levels',
    yaxis_title='Number of Answers',
    barmode='group',
    hovermode='x unified'
)

# Show the figure
fig_difficulty.show()
```
